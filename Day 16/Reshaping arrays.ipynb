```python
#Reshaping Arrays
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.shape)
```

```python
#Convert 1D â†’ 2D
reshaped = arr.reshape(2, 3)
print(reshaped)


Output:

[[1 2 3]
 [4 5 6]]


âš  Rule:
Total elements must match.
2 Ã— 3 = 6 âœ”

You can also use -1:

arr.reshape(3, -1)


NumPy automatically calculates the remaining dimension.

ðŸ”¹ flatten()

Converts multi-dimensional array into 1D copy.

arr2 = np.array([[1,2,3],[4,5,6]])
flat = arr2.flatten()
print(flat)


Output:

[1 2 3 4 5 6]


âœ” Returns a copy (changes wonâ€™t affect original)

ðŸ”¹ ravel()

Also converts to 1D.

r = arr2.ravel()
print(r)


âœ” Returns a view (not copy)

If you modify r, original may change.
```
