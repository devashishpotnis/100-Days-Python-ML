```python
import numpy as np
```

```python
#Element-wise Operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

a + b    # [5 7 9]
a - b    # [-3 -3 -3]
a * b    # [4 10 18]
a / b    # [0.25 0.4 0.5]
```

```python
#Scalar Operations (Broadcasting)
a + 10   # [11 12 13]
a * 2    # [2 4 6]
```

```python
#NumPy Math Functions
np.sum(a)     # 6
np.mean(a)    # 2.0
np.max(a)     # 3
np.min(a)     # 1
np.sqrt(a)    # [1. , 1.41, 1.73]
np.exp(a)     # e^x

```

```python
#Operations on 2D Arrays
arr = np.array([[1, 2], [3, 4]])

arr.sum()        # 10
arr.sum(axis=0)  # Column-wise → [4 6]
arr.sum(axis=1)  # Row-wise → [3 7]
```
