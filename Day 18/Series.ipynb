```python
import pandas as pd
```

```python
#Creating a Series
# Creating a Series from a list
numbers = [10, 20, 30, 40, 50]

series = pd.Series(numbers)

print("Series:")
print(series)
```

```python
#Series with custom index
series2 = pd.Series(numbers, index=["a", "b", "c", "d", "e"])

print("\nSeries with custom index:")
print(series2)
```
